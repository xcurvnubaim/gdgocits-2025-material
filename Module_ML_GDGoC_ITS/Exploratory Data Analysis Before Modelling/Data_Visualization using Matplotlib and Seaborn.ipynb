## Jenis Analisis Visualisasi Data

Bergantung pada jumlah variabel yang ingin dipelajari sekaligus, ada tiga jenis analisis visualisasi data.

- Analisis univariat:  Digunakan untuk meringkas perilaku hanya satu variabel pada satu waktu.
- Analisis bivariat:  Membantu mempelajari hubungan antara dua variabel
- Analisis multivariat:  Memungkinkan praktisi data menganalisis lebih dari dua variabel sekaligus.

3. Histogram

Histogram menunjukkan distribusi variabel numerik dengan batangan. Untuk membuat histogram, mula-mula data numerik dibagi ke dalam beberapa rentang atau bin, dan frekuensi kemunculan setiap rentang dihitung. Sumbu horizontal menunjukkan rentang, sedangkan sumbu vertikal menunjukkan frekuensi atau persentase kemunculan suatu rentang.

Source : (Datacamp)

4. Boxplot


Selain histogram, plot lain yang bagus untuk menunjukkan distribusi variabel atau data adalah boxplot. Boxplot menyediakan cara yang intuitif dan menarik untuk menemukan elemen-elemen berikut:

- Median . Nilai tengah dari kumpulan data di mana 50% datanya kurang dari median dan 50% datanya lebih tinggi dari median.
- Kuartil atas . Persentil ke-75 dari kumpulan data yang 75% datanya lebih kecil dari kuartil atas, dan 25% datanya lebih besar dari kuartil atas.
- Kuartil bawah . Persentil ke-25 dari kumpulan data yang 25% datanya lebih kecil dari kuartil bawah dan 75% lebih tinggi dari kuartil bawah.
- Rentang interkuartil . Kuartil atas dikurangi kuartil bawah
- Nilai yang berdekatan di atas . Atau dalam bahasa sehari-hari disebut "maksimum". Nilai ini mewakili kuartil atas ditambah 1,5 kali rentang interkuartil.
- Nilai yang berdekatan lebih rendah . Atau dalam bahasa sehari-hari disebut "minimum." Nilai ini mewakili kuartil bawah dikurangi 1,5 kali rentang interkuartil.
- Nilai outlier . Nilai apa pun di atas nilai maksimum atau di bawah nilai minimum.

## Visualisasi Univariate

### Data Heart Disease

```python
#Histogram Distribusi BMI
plt.figure(figsize=(6,4))
sns.histplot(heart_df['BMI'], kde=True, bins=30, color='skyblue')
plt.title("Distribusi BMI") #Menambahkan judul pada grafik.
plt.xlabel("BMI") #Menambahkan label pada sumbu x
plt.ylabel("Frekuensi") #Menambahkan label pada sumbu y.
plt.grid(True) #Menambah grid
plt.show() #Menampilkan grafik.
```

`bins`: Menentukan jumlah interval pada histogram.

```python
#Boxplot : BMI berdasarkan Kategori Usia
plt.figure(figsize=(10,5))
sns.boxplot(data=heart_df, x='AgeCategory', y='BMI', palette='pastel')
plt.xticks(rotation=45)
plt.title("BMI berdasarkan Kategori Usia")
plt.xlabel("Kategori Usia")
plt.ylabel("BMI")
plt.show()
```

```python
# Barplot : Distribusi Jenis Kelamin
gender_counts = heart_df['Sex'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title("Distribusi Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Jumlah")
plt.show()
```

**Variasi Grafik Batang**<br>
**Grafik Batang Horizontal**: Gunakan `plt.barh()` untuk menampilkan grafik batang secara horizontal.

```python
plt.barh(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title("Distribusi Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Jumlah")
plt.show()
```

```python
#Pie Chart : Distribusi Jenis Kelamin
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Contoh Diagram Lingkaran")
plt.show()
```

`labels`: Memberi label pada tiap bagian.<br>
`autopct`: Menampilkan persentase tiap bagian.<br>
`startangle`: Mengatur sudut awal diagram.

## Visualisasi Multivariate

### Data Iris

```python
#Heatmap: Korelasi antar fitur
plt.figure(figsize=(6,5))
sns.heatmap(iris_df.drop(columns='class').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Korelasi antar fitur Iris")
plt.show()
```

```python
#Bubble Chart: Panjang vs Lebar Petal
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=iris_df,
    x='petal_length',
    y='petal_width',
    size='sepal_length',
    hue='class',
    palette='viridis',
    alpha=0.7,
    sizes=(20, 200)
)
plt.title("Bubble Chart Petal")
plt.show()
```

```python
#Scatter Plot: BMI vs Physical Health
plt.figure(figsize=(6,4))
sns.scatterplot(data=heart_df, x='BMI', y='PhysicalHealth', hue='HeartDisease', alpha=0.6)
plt.title("BMI vs PhysicalHealth")
plt.xlabel("BMI")
plt.ylabel("Physical Health")
plt.grid(True)
plt.show()
```

```python
#Pairplot: Visualisasi menyeluruh antar fitur
sns.pairplot(iris_df, hue='class', palette='husl')
plt.suptitle("Pairplot Dataset Iris", y=1.02)
plt.show()
```
