```python
#Menggunakan data x
import numpy as np
import statistics

mean1 = sum(x)/len(x)
mean2 = statistics.mean(x)
mean3 = np.mean(x)

# Mencetak semua hasil dalam satu print statement
print(f"Mean1 (menggunakan sum dan len): {mean1}")
print(f"Mean2 (menggunakan statistics.mean): {mean2}")
print(f"Mean3 (menggunakan numpy.mean): {mean3}")
```

Varian

```python
#Menggunakan data x
n = len(x)
mean_= sum(x)/len(x)
var1 = sum((item - mean_)**2 for item in x) / (n-1)
var2 = statistics.variance(x)
var3 = np.var(x)

# Mencetak semua hasil dalam satu print statement
print(f"Varian1 (menggunakan sum dan len): {var1}")
print(f"Varian2 (menggunakan statistics.variance): {var2}")
print(f"Varian3 (menggunakan numpy.var): {var3}")
```

```python
#Menggunakan data x_with_nan
n = len(x_with_nan)
mean_ = sum(x_with_nan)/len(x_with_nan)
var1 = sum((item - mean_)**2 for item in x_with_nan) / (n-1)
var2 = np.var(x_with_nan)

# Mencetak semua hasil dalam satu print statement
print(f"Varian1 (menggunakan sum dan len): {var1}")
print(f"Varian2 (menggunakan numpy.var): {var2}")

#tidak bisa menggunakan statistics karena cannot convert NaN to integer ratio
```

Standard Deviasi

```python
mean_=sum(x)/len(x)
var_ = sum((item - mean_)**2 for item in x) / (len(x)-1)
std_ = var_ ** 0.5
print(std_)
print(np.std(x))
print(statistics.stdev(x))
```

## Mengelompokkan Data

### Dasar Penggunaan apply()

Fungsi `apply()` adalah salah satu fungsi paling fleksibel dan kuat di Pandas. Fungsi ini memungkinkan pengguna menerapkan fungsi pada setiap elemen dalam DataFrame atau Series. Dengan `apply()`, Anda dapat melakukan transformasi data, perhitungan, atau bahkan mengubah bentuk data secara cepat dan efisien.

Parameter `axis` menentukan arah penerapan fungsi:

- `axis=0` (Default): Fungsi diterapkan pada kolom (beroperasi antar baris).
- `axis=1`: Fungsi diterapkan pada baris (beroperasi antar kolom).


```python
# Define a function to categorize ages
def categorize_age(age):
  if age < 18:
    return 'Child'
  elif age < 60:
    return 'Adult'
  else:
    return 'Senior'

# Apply the function to the 'age' column
df['age_category'] = df['age'].apply(categorize_age)

# Display the DataFrame with the new 'age_category' column
df
```

### Dasar Penggunaan groupby()

Anda dapat mengelompokkan data dan melakukan agregasi menggunakan `groupby`.

```python
# Group data by 'sex' and calculate the mean of 'age' for each group
grouped_data = df.groupby('sex')['age'].mean()
grouped_data
```
